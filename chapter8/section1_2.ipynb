# インデックス、スライシング、結合



## 指定された次元の指定されたシーケンスTensor同士を連結する

  


```
x = torch.randn(2, 3)
x
```

```
torch.cat((x, x, x), 0)
```

```
torch.cat((x, x, x), 1)
```

## インデックスを指定して既存のTensorからTensorを切り出す


```
x = torch.randn(3, 4)
x
```

```
x[1]
```

```
x[:,0]
```

```
indices = torch.tensor([0, 2])
torch.index_select(x, 0, indices)
```

```
torch.index_select(x, 1, indices)
```

## 大きさが１の次元を削除したTensorを作成する


```
x = torch.zeros(2, 1, 2, 1, 2)
x.size()
```

```
y = torch.squeeze(x)
y.size()
```

```
y = torch.squeeze(x, 0)
y.size()
```

```
y = torch.squeeze(x, 1)
y.size()
```

```
y = torch.squeeze(x, 3)
y.size()
```

## Tensorを転置する


```
x = torch.randn(2, 3)
x
```

```
torch.t(x)
```

## Tensorの次元を入れ替える



```
x = torch.randn(2, 3)
x
```

```
torch.transpose(x, 0, 1)
```
