## 埋め込みレイヤー

```
# 10テンソルのサイズ3を含む埋め込みモジュール
embedding = nn.Embedding(10, 3)
# それぞれ4つのインデックスの2つのサンプルのバッチ
input = torch.LongTensor([[1,2,4,5],[4,3,2,9]])
embedding(input)
```

```
# padding_idxの例
embedding = nn.Embedding(10, 3, padding_idx=0)
input = torch.LongTensor([[0,2,0,5]])
embedding(input)
```
