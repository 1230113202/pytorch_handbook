# 色々なリカレントニューラルネット

## 多対１のリカレントニューラルネット

```
import torch
import torch.nn as nn

batch_size = 20
time_steps = 10
input_size = 2
num_classes = 2

net = nn.LSTM(input_size, num_classes, num_layers=2, batch_first=True)
input_seq = torch.randn(batch_size, time_steps, input_size)
print("input_seq：", input_seq.size()) #10シーケンスの2次元データ、バッチは20。
output_seq, _ = net(input_seq)
out = output_seq[:, -1, :] #スライスで最終のシーケンスのみ使う。
print("output：", out.size())
```

```
import torch
import torch.nn as nn

batch_size = 20
time_steps = 10
input_size = 2
hidden_size = 2
num_classes = 2

net = nn.LSTMCell(input_size, hidden_size)
input_seq = torch.randn(batch_size, time_steps, input_size)
print("input_seq：", input_seq.size()) #10シーケンスの2次元データ、バッチは20。
#隠れ状態の初期化
hx = torch.randn(batch_size, hidden_size)
cx = torch.randn(batch_size, hidden_size)
#10シーケンス分ループします。
for i in range(time_steps):
    hx, cx = net(input_seq[:, -1, :], (hx, cx))
    if i == 9:
        #最終シーケンスの出力値
        print("output：", hx.size()) 
```

# 多対多のリカレントニューラルネット

```
import torch
import torch.nn as nn

batch_size = 20
time_steps = 10
input_size = 2
num_classes = 2

net = nn.LSTM(input_size, num_classes, num_layers=2, batch_first=True)
input_seq = torch.randn(batch_size, time_steps, input_size)
print("input_seq：", input_seq.size()) #10シーケンスの2次元データ、バッチは20。
output_seq, (h_n, c_n) = net(input_seq)
print(output_seq.size())
```
