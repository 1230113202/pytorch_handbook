# 数学演算

## 要素単位の絶対値を計算する



```
x = torch.abs(torch.tensor([-1, -2, 3]))
x
```

## 要素単位でスカラー値を加算する


```
x = torch.randn(4)
x
```

```
x = torch.add(x, 20)
x
```

## Tensorの値を指定範囲に収める


```
x = torch.randn(4)
x
```

```
torch.clamp(x, min=-0.5, max=0.5)
```

```
x = torch.randn(4)
x
```

```
x = torch.clamp(x, max=0.5)
x
```

## 入力の各要素をスカラー値で除算する


```
x = torch.randn(5)
x
```

```
torch.div(x, 0.5)
```

```
x1 = torch.randn(4, 4)
x1
```

```
x2 = torch.randn(4)
x2
```

```
torch.div(x1, x2)
```

## 入力の要素に指数関数を計算する


```
import math
torch.exp(torch.tensor([0, math.log(2)]))
```

## 自然対数を計算する


```
torch.log(torch.randn(5))
```

## 各要素で乗算する


```
x = torch.randn(3)
x
```

```
torch.mul(x, 100)
```

```
x1 = torch.randn(4, 1)
x1
```

```
x2 = torch.randn(1, 4)
x2
```

```
torch.mul(x1, x2)
```

## 最も近い整数に丸める


```
x = torch.randn(4)
x
```

```
torch.round(x)
```

## sigmoidを計算する



```
x = torch.randn(4)
x
```

```
torch.sigmoid(x)
```

## 符号を得る



```
x = torch.randn(4)
x
```

```
torch.sign(x)
```

## 平方根を計算する



```
x = torch.randn(4)
x
```

```
torch.sqrt(x)
```

## tanhを計算する


```
x = torch.randn(4)
x
```

```
torch.tanh(x)
```

## バッチを含んだTensor同士の行列積を計算する



```
batch1 = torch.randn(10, 3, 4)
batch2 = torch.randn(10, 4, 5)
res = torch.bmm(batch1, batch2)
res.size()
```

## １階Tensor同士の内積を計算する




```
torch.dot(torch.tensor([2, 3]), torch.tensor([2, 1]))
```

## 行列Tensor同士の行列積を計算する



```
mat1 = torch.randn(2, 3)
mat2 = torch.randn(3, 3)
torch.mm(mat1, mat2)
```

## 色々なTensor同士の行列積を計算する



```
# ベクトル x ベクトル
tensor1 = torch.randn(3)
tensor2 = torch.randn(3)
torch.matmul(tensor1, tensor2).size()
```

```
# 行列 x ベクトル
tensor1 = torch.randn(3, 4)
tensor2 = torch.randn(4)
torch.matmul(tensor1, tensor2).size()
```

```
# バッチ行列 x ブロードキャストされたベクトル
tensor1 = torch.randn(10, 3, 4)
tensor2 = torch.randn(4)
torch.matmul(tensor1, tensor2).size()
```

```
# バッチ行列 x バッチ行列
tensor1 = torch.randn(10, 3, 4)
tensor2 = torch.randn(10, 4, 5)
torch.matmul(tensor1, tensor2).size()
```

```
#バッチ処理された行列 x ブロードキャストされた行列
tensor1 = torch.randn(10, 3, 4)
tensor2 = torch.randn(4, 5)
torch.matmul(tensor1, tensor2).size()
```
